# Axis

        axis = 0 or 1

axis = 0 is also lnown as axis = "index"

axis = 1 is also known as axis = "columns'

### For mathematical & changing each data operation like mean, median, apply etc,

| Axis     | Moves Across     | Means You’re Acting On |
| -------- | ---------------- | ---------------------- |
| `axis=0` | Down columns (↓) | Each **column**        |
| `axis=1` | Across rows (→)  | Each **row**           |

🎨 Visual Summary:

Axis 0 (↓): Think "columns", move down rows

Axis 1 (→): Think "rows", move across columns

🎈 Trick to Remember:

axis=0 → 0 is the first direction → vertical → like height

axis=1 → 1 is the second direction → horizontal → like width

### For read, write operation like drop, filter etc

| Axis     | Moves Across     | Means You’re Acting On |
| -------- | ---------------- | ---------------------- |
| `axis=0` | Across columns (→) | Each **column**        |
| `axis=1` | Down rows (↓)  | Each **row**           |

🎨 Visual Summary:

Axis 0 (→): Think "rows", move across columns

Axis 1 (↓): Think "columns", move down rows

🎈 Trick to Remember:

axis=0 → 0 is the second direction → horizontal → like width

axis=1 → 1 is the first direction → vertical → like height



Here’s an example DataFrame to help:

            Name	Age	Score
            Amy	10	90
            Bob	12	80
            Cara	11	85

This table has:

Rows: Amy, Bob, Cara

Columns: Name, Age, Score

#### ===> For Mean,

Let’s say we want the average score and age (not name, because it's text):

                df.mean(axis=0)

This means:

"Go down each column and find the average of numbers"

You’ll get:

                Age     11.0
                Score   85.0

Because we said axis=0, pandas looks down each column.


But if you say:

                df.mean(axis=0)

This means:

"Go across each row and average the numbers in that row"

You’ll get:

                0    50.0  # Amy → (10 + 90) / 2
                1    46.0  # Bob → (12 + 80) / 2
                2    48.0  # Cara → (11 + 85) / 2

Because we said axis=1, pandas looks across each row.


### ===> For Drop,

If you write:

                df.drop('Age', axis=1)

You're saying:

“Hey, go across columns and remove the one named ‘Age’”

Output:

                | Name | Score |
                |------|-------|
                | Amy  | 90    |
                | Bob  | 80    |
                | Cara | 85    |

But if we say:

                df.drop(1, axis=0)

You're saying:

“Go down rows and remove the row with index 1 (Bob)”

Output:

                | Name | Age | Score |
                |------|-----|-------|
                | Amy  | 10  | 90    |
                | Cara | 11  | 85    |
