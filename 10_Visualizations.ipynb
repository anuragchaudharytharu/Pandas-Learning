### Visualizations

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv(r"D:\Artificial Intelligence Engineer\Pandas\Pandas Learning\Ice Cream Ratings.csv")
df = df.set_index("Date")
df
```

### plot

                .plot()

##### parameter: 
        df ===> dataframe

        x/y = ".." ===> which columns of a DataFrame to use for the axes of the plot.

        xlabel ===> x-axis label name

        ylabel ===> y-axis label name

        kind = ".." ====> plot type we want. By default, we get line plot if not mentioned in this parameter

        subplots = True/False ====> each column gets separate plot visualizations if True is provided

        title = ".." ====> name of the plot

        stacked = True/False ===> It make traditional bar plot visualization to stacked bar plot visualization


```python
df.plot()
```

```python
# styple of visualizations
print(plt.style.available)
plt.style.use("fivethirtyeight")
```

```python
# what kind of plot visualization we want ====>  eg: pie, bar, line, box, etc
df.plot(kind="box")
```

```python
# supbplots ===> separate visualization for each column datas
df.plot(kind="line", subplots=True)
```

```python
# title of the plot and label name for x-axis & y-axis 
df.plot(kind="line", title="Ice Cream Ratings", xlabel="Daily Rating", ylabel="Scores")
```

```python
# stacked plot visualization ===> specific to bar chart plot only
df.plot(kind="bar", stacked = True)
```

```python
# specific column bar char plot visualization
df["Flavor Rating"].plot(kind="bar")
```

```python
# Horizontal Bar Chart Visualizations 
df.plot.barh(stacked = False)
```

```python
# scatter plot ===> we need to specify the x-axis and y-axis
df.plot.scatter(x = "Texture Rating", y = "Overall Rating")
```

```python
# Size of each point for scatter plot visualization
df.plot.scatter(x = "Texture Rating", y = "Overall Rating", s = 500)
```

```python
# Color of each point for scatter plot visualization
df.plot.scatter(x = "Texture Rating", y = "Overall Rating", s = 500, c="#FF0000")
```

```python
# bins parameter controls how the data is grouped (or binned) along the x-axis.
# if we do bins=5, then it means the data is divided into 5 groups
# Simply its a total number of tick marks present in the x-axis 
df.plot.hist(bins=5)
```

```python
df.plot.hist(bins=10)
```

```python
# boxplot
df.boxplot()
```

```python
# area plot
df.plot.area()
```

```python
# figure size for area plot
df.plot.area(figsize = (10,5))
```

```python
# pie chart with column specified of y-axis and with figure size mentioned. Pie chart only accepts y-axis columns 
df.plot.pie(y="Flavor Rating", figsize = (10,10))
```
