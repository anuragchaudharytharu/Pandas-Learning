```python
# Last_Name column ===> Cleaning the unwanted character from the data

# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.lstrip("..")

df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
# Phone_Number column ===> removing the unwanted character from the data and formatting the data in certain pattern

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
df
```

```python
# Phone_Number column ===> formatting the data in certain phone pattern


df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)) # converting the data to string

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:]) # formatting to phone pattern

df["Phone_Number"] = df["Phone_Number"].str.replace("nan--", "") # removing nan--

df["Phone_Number"] = df["Phone_Number"].str.replace("Na--", "") #removing Na--

df
```

```python
"""
                    df["Address"].astype(str).str.split(",", 2)

    It gives error:
        StringMethods.split() takes from 1 to 2 positional arguments but 3 were given

    This code gives error because python think we are doing:
                    split(arg1=',', arg2=2, arg3=<nothing>)

    But str.split() can only accept:
                    split(pat=None, n=1, expand=False)
                    
                            pat="separator" ===> separator to use for splitting
                            
                            n ===> number of times we want to split
                            
                            expand=True/False ===> expanding each split into separate columns
                            
                            
                    
                    
                    
    
    So we must use pat= and n= explicitly to avoid this TypeError.
"""
# Address column ===> Splitting the columns. First we convert the data to string and then we split into number of desired times
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].astype(str).str.split(pat=",", n=2, expand=True)
df
```

```python
# Paying Customer column and Do_Not_Contact column ===> replacing messy data with proper Yes and No value
df[["Paying Customer", "Do_Not_Contact"]] = df[["Paying Customer", "Do_Not_Contact"]].replace({"Yes": "Y", "No": "N"})
df
```

```python
# Replace all the N/a value in the dataframe with null value
df.replace("N/a", "", inplace=True)
```

```python
# Filling all the null values with empty blank space
df.fillna("", inplace=True)
```

```python
# Removing the row if Do_Not_Contact column consist of value Y 
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == "Y":
        df.drop(x, inplace=True)

df
```

```python
# Removing Phone_Number column with empty blance space data

# df.dropna(subset=["columnName", "columnName", ...], inplace = True)

for x in df.index:
    if df.loc[x, "Phone_Number"] == "":
        df.drop(x, inplace=True)
df
```
