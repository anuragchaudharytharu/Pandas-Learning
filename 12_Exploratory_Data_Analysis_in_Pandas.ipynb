```python
# It groups data by continent, picks certain columns by slicing and sorting their names alphabetically, calculates the average for these columns, rounds the results, and sorts by 2022 population from largest to smallest.
df2 = df.groupby("Continent")[sorted(df.columns[5:len(df.columns)- 4])].mean(numeric_only=True).round(2).sort_values(by="2022 Population", ascending=False)
df2
```

```python
# transposing the datarame ====>  columns to rows and vice versa
df3 = df2.transpose()
df3
```

```python
df3.plot()
```
